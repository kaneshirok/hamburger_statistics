```python
# もぐもぐバーガーのポテトの度数分布
column_name = mogumogu_potato.columns[0]
new_columns_name = ('0cm以上1cm未満', '1cm以上2cm未満','2cm以上3cm未満', '3cm以上4cm未満', '4cm以上5cm未満', '5cm以上6cm未満', '6cm以上7cm未満', '7cm以上8cm未満')
length_0_1 = mogumogu_potato[mogumogu_potato[column_name] < 1].rename(columns={column_name: new_columns_name[0]})[new_columns_name[0]]
length_0_1 = pd.DataFrame(pd.DataFrame(length_0_1).count())
length_1_2 = mogumogu_potato[(1 <= mogumogu_potato[column_name]) & (mogumogu_potato[column_name] < 2)].rename(columns={column_name: new_columns_name[1]})[new_columns_name[1]]
length_1_2 = pd.DataFrame(pd.DataFrame(length_1_2).count())
length_2_3 = mogumogu_potato[(2 <= mogumogu_potato[column_name]) & (mogumogu_potato[column_name] < 3)].rename(columns={column_name: new_columns_name[2]})[new_columns_name[2]]
length_2_3 = pd.DataFrame(pd.DataFrame(length_2_3).count())
length_3_4 = mogumogu_potato[(3 <= mogumogu_potato[column_name]) & (mogumogu_potato[column_name] < 4)].rename(columns={column_name: new_columns_name[3]})[new_columns_name[3]]
length_3_4 = pd.DataFrame(pd.DataFrame(length_3_4).count())
length_4_5 = mogumogu_potato[(4 <= mogumogu_potato[column_name]) & (mogumogu_potato[column_name] < 5)].rename(columns={column_name: new_columns_name[4]})[new_columns_name[4]]
length_4_5 = pd.DataFrame(pd.DataFrame(length_4_5).count())
length_5_6 = mogumogu_potato[(5 <= mogumogu_potato[column_name]) & (mogumogu_potato[column_name] < 6)].rename(columns={column_name: new_columns_name[5]})[new_columns_name[5]]
length_5_6 = pd.DataFrame(pd.DataFrame(length_5_6).count())
length_6_7 = mogumogu_potato[(6 <= mogumogu_potato[column_name]) & (mogumogu_potato[column_name] < 7)].rename(columns={column_name: new_columns_name[6]})[new_columns_name[6]]
length_6_7 = pd.DataFrame(pd.DataFrame(length_6_7).count())
length_7_8 = mogumogu_potato[(7 <= mogumogu_potato[column_name]) & (mogumogu_potato[column_name] < 8)].rename(columns={column_name: new_columns_name[7]})[new_columns_name[7]]
length_7_8 = pd.DataFrame(pd.DataFrame(length_7_8).count())
mogumogu_fd = pd.concat([length_0_1, length_1_2, length_2_3, length_3_4, length_4_5, length_5_6, length_6_7, length_7_8])
mogumogu_fd = mogumogu_fd.rename(columns={0: 'ポテトの長さ'})
mogumogu_fd.hist(bins=1)
plt.show()
```
